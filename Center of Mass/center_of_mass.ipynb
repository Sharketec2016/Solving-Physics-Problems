```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sm
plt.style.use('classic')
```

Within this file I code up a center of mass matrix of a system. 

```python
#first we look at our center of mass matrix. This will be a matrix made up of 3 coordinate points. For our sake we are going to limit it to 2-D for simplicity

R = np.array([]) #center of mass matrix

#lets now define our given points of the system
given_masses = np.array([5, 2, 3, 7, 7, 1]) #units of kg
given_x = np.array([2, 8, 5, 10, 3, 6])
given_y = np.array([6, 2, 11, 5, 4, 9])

R_x = np.dot(given_masses, given_x)/np.sum(given_masses)
R_y = np.dot(given_masses, given_y)/np.sum(given_masses)

plt.scatter(given_x, given_y, marker='o', color='black', s=50)
plt.scatter(R_x, R_y, marker='*', s=80)
plt.grid()
```

```python
given_x = np.random.randint(1, 11, size=10)
given_y = np.random.randint(1, 11, size=10)
given_masses = np.random.randint(1, 11, size=10)

R_x = np.dot(given_masses, given_x)/np.sum(given_masses)
R_y = np.dot(given_masses, given_y)/np.sum(given_masses)

plt.scatter(given_x, given_y, marker='o', color='black', s=50)
plt.scatter(R_x, R_y, marker='*', s=80)
plt.text(R_x+0.2, R_y+0.2, "COM")
plt.grid()
```
