##### Scalar products, when you scale a vector by some quantity. Let $\nu$ be some vector [3, 4], then:
$$ \nu=[3,4] \\ 5\nu=5*[3,4]=[5*3, 5*4]=[15, 20]$$

```python
v = np.array([[3, 4]])
5*v
```

#### Dot Products, when you multiply two vectors together. The result of a dot product is a scalar.
$$ \vec{a}=[a_x,a_y,a_z] \\ \vec{b}=[b_x,b_y,b_z] \\ \vec{a}\cdot\vec{b} = a_x \cdot b_x + a_y \cdot b_y + a_z \cdot b_z OR \\ \vec{a}\cdot\vec{b} = |\vec{a}||\vec{b}|cos(\theta)$$

```python
a = np.array([2, 4, 6])
b = np.array([1, 2, 3])

np.dot(a, b)
```

#### Cross Products, when you cross two vectors together. The result of a cross product is another vector that is perpendicular to the two inital vectors

$$ \vec{a}=[a_x,a_y,a_z] \\ \vec{b}=[b_x,b_y,b_z] \\ \vec{a} \times \vec{b} = 
\begin{vmatrix}
x & y & z\\
a_x & a_y & a_z\\
b_x & b_y & b_z\\
\end{vmatrix} OR \\
\vec{a} \times \vec{b} = |\vec{a}||\vec{b}|sin(\theta)
$$

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
np.cross(a, b)
```

#### The angle between two vectors $\theta$ can be related through the angle between their resultant vector $\phi$:
> $$ \tan{\phi}=\frac{\boldsymbol{a}\sin(\theta)}{\boldsymbol{b}+\boldsymbol{a}\cos(\theta)} $$
#### As shown above from the figure, the magnitude of a vector is described through its compnents using Pythagorems Theorem.
>$$ \boldsymbol{a}=|\vec{a}|=\sqrt{\vec{b}^2+\vec{c}^2} $$
#### Sometimes, when necessary, it is easier to describe our system using unit vectors. Unit vectors are vectors that point in the same direction as our vector of interest, but have been normalized. Normalization means that each component of our vector has been scaled down such that the sum of its components equals 1. The notation for a unit vector is $\hat{a}$.
> $$ \hat{a} = \frac{\vec{a}}{|\vec{a}|} = (\frac{a_x}{\sqrt{a_x^2+a_y^2+a_z^2}}\hat{i}, \frac{a_y}{\sqrt{a_x^2+a_y^2+a_z^2}}\hat{j}, \frac{a_z}{\sqrt{a_x^2+a_y^2+a_z^2}}\hat{k}) \\ 
\frac{a_x}{\sqrt{a_x^2+a_y^2+a_z^2}}\hat{i} + \frac{a_y}{\sqrt{a_x^2+a_y^2+a_z^2}}\hat{j} + \frac{a_z}{\sqrt{a_x^2+a_y^2+a_z^2}}\hat{k} = 1
$$

# Motion in Two and Three Dimensions
##### We are now going to apply the information above, and some stuff from trig, to look at some concepts of motion within two and three dimesions.

##### First, a particle moving within a medium will follow a path or trajectory. This path is described within 3-D as:
> $$ Position:\hspace{1em}\vec{r}=x(t)\hat{i}+y(t)\hat{j}+z(t)\hat{k}$$
##### Vector $\vec{r}$ is known as the position of the object. If we wanted to know the displacement vector $\Delta\vec{r}$, we would find the position at two different points in time, $t$ and $t+\Delta t$, and evaluate:
> $$ Displacement:\hspace{1em}\Delta \vec{r} = \vec{r}_{t+\Delta t} - \vec{r}_{t}$$
##### For more detail about different equations of (projectile) motion, check folder on throwing objects. Past the position vector, we can use this to define the instantaneos velocity of a object at some point $\vec{r}(t)$ as:
> $$ Intantaneous\hspace{0.5em}Velocity:\hspace{1em}\hspace{1em}\vec{\nu}(t) = \lim_{\Delta t \rightarrow 0}\frac{\vec{r}(t+\Delta t)-\vec{r(t)}}{\Delta t}=\frac{d\vec{r}}{dt}$$
##### The __Instantaneous__ part when measuring the velocity is seen within the $\lim_{\Delta t \rightarrow 0}$. As the change in time $\Delta t$ becomes __theoretically__ infinitly small, we evaluate the function at essentially one point. Thus at that one point, we find what the velocity of our object is. Velocity, similar to position $\vec{r}$, is also a vector. Meaning that it has positional components, appearing as:
> $$\vec{\nu} = (\nu_x, \nu_y, \nu_z) $$
##### To find the magnitude of a vector, we follow the description of a vector above, and get:
> $$ Magnitude:\hspace{1em}\boldsymbol{\nu}=|\vec{\nu}|=\sqrt{\nu_x^2+\nu_y^2+\nu_z^2} $$
#### With this information we can now extend it to acceleration $\vec{a}$:
> $$ Acceleration:\hspace{1em}\vec{a}(t) = \lim_{\Delta t \rightarrow 0}\frac{\vec{\nu}(t+\Delta t)-\vec{\nu(t)}}{\Delta t}=\frac{d\vec{\nu}}{dt}$$
#### Using this same understanding, we will now extend our conecepts of translational motion into rotational motion. Rotational motion consists of similar quantities as translational, such as velocity and acceleration, but pertain to bodies moving in a curved path/rotational motion.
> $$ Angular\hspace{0.5em}Velocity:\hspace{1em}\omega=\frac{d\phi}{dt} $$
#### We relate translational and angular velocity together through:
> $$ \boldsymbol{\nu}=\omega R$$
#### Where R is the Range function, or the radius of the circular path the object takes. $\omega$ has another expression interms of its angular frequency $f$, and time period T:
> $$ \omega=2\pi f \hspace{1em}\&\hspace{1em}f=\frac{1}{T} $$
#### T is the time it take for our system to start repeating itself. Think of this as the time it takes for the object to go around a circle once. After going around the circle once, it starts repeating its path and becomes indistinguishable from any other time at that position.

# Problem 3
#### The motion of a planet about a star is described by the vector $\vec{r}_1=R\cos{(2\pi ft)}\hat{i}+R\sin{(2\pi ft)}\hat{j}$. The position of another planet about the same star is $\vec{r}_2=4R\cos{(\pi ft/4)}\hat{i}+4R\sin{(\pi ft/4)}\hat{j}$. 
> 1) Show that both of these describe circular motion with constant speed. 
> 2) Calculate the position vector of the second planet relative to the first planet. 
> 3) Sketch the path by noting that the speed of one of the planets in its circular motion is eight times larger than the speed of the other.

```python
#part one we can sovle by defining our variables, and then plotting the function
def r1(R, f, t):
    return R*np.cos(2*np.pi*f*t/180) + R*np.sin(2*np.pi*f*t/180) #here within the function definitions, i have to divide by 180 degrees as the numpy functions only take units of radians within the function definition. Thus a division of 180 is necessary
def r2(R, f, t):
    return 4*R*np.cos((np.pi*f*t)/4/180) + 4*R*np.sin((np.pi*f*t)/4/180)

#since we are assuming circular motion, R is constant. I will define R to be 1 to simplify things. Frequency is also assumed to be constant, as we are not told of any damping forces to consider. Thus the only variable that will vary is time t
R = 100
f = 1
t = np.linspace(0, 360, 3600) #im making the size of t to be 3600 to have many points drawing the different orbits



#next we are asked to calculate the position vector of one planet w.r.t the other planet. Thus we will be finding the displacement vector between r2 and r1. For sake of comparision, lets find the displacement at time t=0 or any arbitrary time.
R_sm, f_sm, t_sm = sm.symbols('R f, t')
r1_sm_x = R_sm*sm.cos(2*sm.pi*f_sm*t_sm/180)
r1_sm_y = R_sm*sm.sin(2*sm.pi*f_sm*t_sm/180)
r1_sm = sm.Matrix([r1_sm_x, r1_sm_y])

r2_sm_x = 4*R_sm*sm.cos(sm.pi*f_sm*t_sm/4/180)
r2_sm_y = 4*R_sm*sm.sin(sm.pi*f_sm*t_sm/4/180)
r2_sm = sm.Matrix([r2_sm_x, r2_sm_y])

#for part a, we can show that they are moving at a constant speed by taking the second derivative w.r.t time and show that it should be equal to zero
r1_acc = sm.diff(r1_sm, t_sm, 2) #by printing out this variable here, we would see that each term is divided by 8100. With a division this large, the acceleration would be so small that it could be neglected, and thus is practically zero. In reality, there are gravitational forced from each planets on each other, and the parent star on each planet. Thus the planets will always feel a force and subsequently will always have some non-zero acceleration. Note this opperation is the same for r2_acc, but the denominator is orders of magnitude larger making it even more negligible.
r2_acc = sm.diff(r2_sm, t_sm, 2)


#For part b we first compute the displacement vector, then turn it into a computable array for plotting
dis_r = sm.simplify(r2_sm - r1_sm)
dis_sm_lamb = sm.lambdify([R_sm, f_sm, t_sm], dis_r, 'numpy')
dis_sm_lamb_x = np.array(dis_sm_lamb(R, f, t)[0][0])
dis_sm_lamb_y = np.array(dis_sm_lamb(R, f, t)[1][0])

new_dis_sm_lamb = dis_sm_lamb_x + dis_sm_lamb_y #this variable new_dis_sm_lamb is the displacement vector between the two planets


#Here we are now plotting our answers from above
fig = plt.figure(figsize=(20, 10))
ax1 = fig.add_subplot(121, projection='polar')
ax2 = fig.add_subplot(122)#, projection='polar')
ax1.plot(t, np.abs(r1(R, f, t)), label=r'$\vec{r}_1$')
ax1.plot(t, np.abs(r2(R, f, t)), label=r'$\vec{r}_2$')
ax1.grid()
ax1.legend(fontsize=15)
ax1.set_title("Orbits of the different planets relative to each other")

ax2.plot(t, new_dis_sm_lamb)
ax2.set_title("Displacment vector plotting over time through the orbit")
ax2.grid()
ax2.set_xlabel("Time [s]")
ax2.set_ylabel("Distance [km]")



```

# Equations of Motion
#### To solve furture questions, we are going to take our knowledge of vectors and apply them to different kinematic equations of motion. There are a bunch, but a few we should be familiar with as the rest are in some way derived from these. We will only be defining them in 2-D and negating any problems that explore a 3-D system for right now.
> ### Kinematic Eq's for Constant Acceleration in 2-D
>> $$ \vec{\nu}=(\nu_{x0}+a_xt,\hspace{1em}\nu_{y0}+a_yt) \\
    d=(x_0+\nu_{x0}t+\frac{1}{2}a_xt^2,\hspace{1em}y_0+\nu_{y0}t+\frac{1}{2}a_yt^2)\\
    \nu^2=(\nu_{x0}^2+2a_x(x-x_0),\hspace{1em}\nu_{y0}^2+2a_y(y-y_0))
    $$
> ### Kinematic Eq's for Projectile Motion
>> $$ \vec{\nu}=(\nu_{x0},\hspace{1em}\nu_{y0}-gt) \\
    d=(x_0+\nu_{x0}t,\hspace{1em}y_0+\nu_{y0}t-\frac{1}{2}gt^2)\\
    Range\hspace{0.5em}R=\frac{\nu_0^2\sin(2\theta_0)}{g}
$$

```python
def d_x(x0, v0_x, t):
    return x0 + v0_x*t
def d_y(y0, v0_y, t, g):
    return y0 + v0_y*t - 0.5*g*t**2
def v_y(v0_y, g, t):
    return v0_y - g*t
def v_y2(v0_y, g, y, y0, t):
    return v0_y**2 - 2*g*(d_y(y0, v0_y, t, g) - y0)
def Range(v0, theta, g):
    return (v0**2 * np.sin(2*theta))/g #remember that this angle theta has to be in radians. 
```

$$  \ddot{x} = -\frac{qB_0}{m}\dot{z} \\ \ddot{z} = \frac{qB_0}{m}\dot{x}  $$

$$ \begin{align}
